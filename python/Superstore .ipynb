```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)
```

```python
# Performance Trend 
performance= pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/Superstore Sales Dataset/data/sql_outputs/overall_performance.csv')
```

```python
performance2=pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/Superstore Sales Dataset/data/sql_outputs/Sales_and_Profit_Trend_per_Year.csv')
```

```python
performance.isna().sum()
performance2.isna().sum()
```

```python
plt.figure()
sns.lineplot(data=performance2,x='Year',y='Total_Sales')
sns.lineplot(data=performance2,x='Year',y='Total_Profit')
plt.title('Sales and Profit over Years')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.savefig('Sales and Profit over Years.png',bbox_inches='tight')
plt.show()
```

```python
# Top Products by Sales
top_products=pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/Superstore Sales Dataset/data/sql_outputs/Top 10 best-selling products.csv')
top_products
```

```python
# Categories Performance
Category=pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/Superstore Sales Dataset/data/sql_outputs/Sales and Profit by Category.csv')
Category
```
